```python
# script to download ghost face images from Google Images
import os
import time
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python

QUERY = 'scary clown'
NUM_IMAGES = 450
OUTPUT_DIR = 'ghost_faces'
DRIVER_PATH = 'path/to/chromedriver'  # Update this to your WebDriver path

# Create output directory if it doesn't exist
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

def fetch_image_urls(query, num_images, driver_path):
    driver = webdriver.Chrome()
    search_url = f"https://www.google.com/search?q={query}&tbm=isch"
    driver.get(search_url)
    image_urls = set()
    while len(image_urls) < num_images:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        for img in soup.find_all('img'):
            try:
                url = img['src']
                if url and url.startswith('http'):
                    image_urls.add(url)
                    if len(image_urls) >= num_images:
                        break
            except KeyError:
                continue
        print(f"Found {len(image_urls)} image URLs so far")
    driver.quit()
    return list(image_urls)

if __name__ == "__main__":
    image_urls = fetch_image_urls(QUERY, NUM_IMAGES, DRIVER_PATH)
```

```python
def download_images(image_urls, output_dir):
    for i, url in enumerate(image_urls):
        try:
            response = requests.get(url)
            with open(os.path.join(output_dir, f'ghost_face_{i+1}.jpg'), 'wb') as f:
                f.write(response.content)
        except Exception as e:
            print(f"Failed to download {url}: {e}")
```

```python
download_images(image_urls, OUTPUT_DIR)
```
